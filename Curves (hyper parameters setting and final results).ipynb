```python
import json
import os.path
import pandas as pd
import holoviews as hv
import json
import numpy as np
hv.extension('bokeh')
%opts Curve [height=500 width=800]
from IPython.core.display import HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
# load the results to draw the curve of the score, each 100 iterations, averaged over the last 100 iterations
with open('hyper_parameters_tuning.json') as json_file:
    results = json.load(json_file)

df = pd.DataFrame(results) 
df['iteration'] = df.index
df.head()

buffer_size = [col for col in df.columns if col.startswith('buffer_size')]
batch_size = [col for col in df.columns if col.startswith('batch_size')]
gamma = [col for col in df.columns if col.startswith('(gamma')]
tau = [col for col in df.columns if col.startswith('(tau')]
network_structure = [col for col in df.columns if col.startswith('(fc1,fc2)')]
lr = [col for col in df.columns if col.startswith('lr_actor, lr_critic')]
noises_gaussian = [col for col in df.columns if col.startswith('noise_GaussianNoise')]
noises_ou = [col for col in df.columns if col.startswith('noise_OUNoise')]
noise_factor_decay = [col for col in df.columns if col.startswith('noise_factor_decay')]
update_rate_update_count = [col for col in df.columns if col.startswith('(update_rate, update_count)')]
inbalanced_replay_memory_positive_reward_ratio = [col for col in df.columns if col.startswith('inbalanced_replay_memory_positive_reward_ratio')]

indicators = {
              'buffer_size' : buffer_size,
              'batch_size' : batch_size,
              '(gamma_start, gamma_end, gamma_increase)' : gamma,
              '(tau_start, tau_end, tau_decay)' : tau,
              '(fc1,fc2)' : network_structure,
              'lr_actor, lr_critic' : lr,
              'noise_GaussianNoise' : noises_gaussian,
              'noise_OUNoise' : noises_ou,
              'noise_factor_decay' : noise_factor_decay,
              '(update_rate, update_count)' : update_rate_update_count,
              'inbalanced_replay_memory_positive_reward_ratio' : inbalanced_replay_memory_positive_reward_ratio
    }
df2 = df.rolling(100).mean()
df2['iteration'] = df2.index
```

```python
global_c = None
for name, indicator in indicators.items() :
    c = None
    for col in indicator :
        if col in df2.columns:
            
            label = col.replace(f'{name}_', "").replace('GaussianNoise', 'gaussian').replace('OUNoise', 'OU')
            current_curve = hv.Curve(df2,  kdims = 'iteration', vdims = col, label=label)
            c = current_curve if c is None else c * current_curve

    if c is not None :
        c = c.opts(legend_position='bottom_right', title = name, xlabel='iterations', ylabel='Score', show_grid=True, height = 480, width = 500)
        global_c = c if global_c is None else global_c + c

display(global_c.cols(3))
```

```python
with open('final_scores.json') as json_file:
    results = json.load(json_file)

df = pd.DataFrame(results) 
df['iteration'] = df.index
df.head()

display(hv.Curve(df,  kdims = 'iteration', vdims = ('best', 'score'), label=label).opts(title = "Score - best parameters"))

df2 = df.rolling(100).mean()
df2['iteration'] = df2.index

display(hv.Curve(df2,  kdims = 'iteration', vdims = ('best', 'score'), label=label).opts(title = "Score over previous 100 - best parameters"))
```
